```python
# TASK 5 - EDA on Titanic Dataset

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Set visual style
sns.set_style('darkgrid')
sns.set_palette('pastel')
plt.rcParams['figure.figsize'] = (10,6)


titanic = pd.read_csv('/content/Titanic-Dataset.csv')

# Preview dataset
titanic.head()
```

```python
# Shape of dataset
print("Shape of dataset:", titanic.shape)

# Info about data types and null values
titanic.info()

# Statistical Summary
titanic.describe(include='all')

# Checking missing values
titanic.isnull().sum()


# 1. Distribution of Age
plt.figure(figsize=(10,6))
sns.histplot(titanic['Age'], kde=True, color='skyblue', bins=30)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()

print("Observation: Age distribution is right-skewed; most passengers are between 20-40 years old.")

# 2. Survival Count
sns.countplot(x='Survived', data=titanic)
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()

print("Observation: More passengers did not survive than survived.")

# 3. Gender Distribution
sns.countplot(x='Sex', data=titanic)
plt.title('Gender Distribution')
plt.xlabel('Sex')
plt.ylabel('Count')
plt.show()

print("Observation: More males onboard than females.")

# 4. Passenger Class Distribution
sns.countplot(x='Pclass', data=titanic)
plt.title('Passenger Class Distribution')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()

print("Observation: Majority of passengers belonged to 3rd class.")
# 5. Survival Rate by Gender
sns.barplot(x='Sex', y='Survived', data=titanic)
plt.title('Survival Rate by Gender')
plt.ylabel('Survival Rate')
plt.show()

print("Observation: Females had a higher survival rate compared to males.")

# 6. Age vs Survival
plt.figure(figsize=(10,6))
sns.kdeplot(data=titanic, x='Age', hue='Survived', fill=True)
plt.title('Age vs Survival')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()

print("Observation: Young children had a slightly higher survival rate.")

# 7. Heatmap of Correlations
plt.figure(figsize=(12,8))
sns.heatmap(titanic.corr(numeric_only=True), annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()

print("Observation: Survival is positively correlated with being female and negatively with being male.")

# 8. Pairplot
sns.pairplot(titanic[['Age','Fare','Pclass','Survived']], hue='Survived', palette='husl')
plt.suptitle('Pairplot Analysis', y=1.02)
plt.show()

print("Observation: Higher fare and lower passenger class (1st class) passengers had higher survival rates.")

print("""
Summary of EDA Findings:
- Majority of passengers were between 20-40 years old.
- More passengers did not survive (about 62%).
- Males were the majority but had lower survival rates compared to females.
- Passengers from 1st class had better chances of survival.
- Younger passengers (children) had slightly better survival odds.
- Fare and Pclass show strong relation with survival.
""")
```
